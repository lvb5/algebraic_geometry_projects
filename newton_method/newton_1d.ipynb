## Finding Roots of Univiarite Polynomial Equations using Newton Method

We code a implementation of the Newton Method below for the 1D cases. The mathematical form used is given by 

$$
x_{n + 1} = x_n - \frac{f(x_n)}{f'(x_n)}. 
$$

```julia
using Calculus
using ForwardDiff
using Plots, ImplicitPlots
```

```julia
function newton_1d(StartPoint::Float64, Func::Function)
    
    start = StartPoint
    tol = Func(start)
    steps = [start]

    i = 0
    while (abs(tol) > 1e-15 && i < 501)
        #test for convergence
        if i == 500
            error("Algorithm did not converge!")
        end

        #implement method
        start = start - Func(start) / derivative(Func, start)
        tol = Func(start)

        #append steps
        append!(steps, start)
        i += 1
    end
    
    return start, steps
end
```

## Some examples

We show a few examples using this function

```julia
f(x) = 3x + 2
root, steps = newton_1d(5.0, f)
plot(f, -5, 5, label = "f(x)")
scatter!([steps], [f.(steps)], label = "steps")
scatter!([root], [f(root)], label = "solution = $root")
```

```julia
f(x) = x^7 + 3x^5 + 5
root, steps = newton_1d(2.0, f)
plot(f, -20, 20, label = "f(x)")
scatter!([steps], [f.(steps)], label = "steps")
scatter!([root], [f(root)], label = "solution = $root")
```

```julia
f(x) = sin(x)
root, steps = newton_1d(2.0, f)
plot(f, -5, 5, label = "f(x)")
scatter!([steps], [f.(steps)], label = "steps")
scatter!([root], [f(root)], label = "solution = $root")
```
